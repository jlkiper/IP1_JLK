# get_data()
This function will accept a filename and then implement the 
code to open and read the parking.csv file into a list.
This should then return the list.

```python
def get_data(filename):
    
    with open(filename, newline='') as f:
        reader = csv.reader(f)                 # Reads file line by line 
        data = list(reader)
    print(data)
    
data = get_data('parking.csv')
```

# show_table()
This function accepts a list and prints a PrettyTable
of the data with the first row of the data as the header row.

```python
def show_table(data):
    
    from prettytable import from_csv
    
    with open("parking.csv", newline = '') as f:     # Formats the table for the csv file
        my_table = from_csv(f)
    print(my_table)

show_table(data)
```

# count_lot_types()
This function accepts a list and the lot type (SURFACE LOT/GARAGE)
and that returns a tuple that contains the lot type and the number
of lots of that type.

```python
def count_lot_type(data, lot_type):
    
    from collections import Counter
    
    with open('parking.csv', newline='') as f:
        my_reader = csv.reader(f, delimiter = ',')                 # Reads csv file
        next(my_reader)                                            # Skips header on count
        data = [row[1] for row in my_reader if row[1] == lot_type] # Reads on the first column
        counter = Counter(data)                                    # Counts the results
        countLot = counter[lot_type]
        return (lot_type, countLot)
    
count_lot_type(data,'SURFACE LOT')
```

# get_column_stats()
This function accepts a list and a column position and returns the Max, Min, and Average for that column.

```python
def get_column_stats(data,column):
    
    from prettytable import PrettyTable
    
    with open("parking.csv", 'r') as f:
        my_reader = list(csv.reader(f, delimiter = ','))
        data = np.array(my_reader)                             # Creates the numpy array from the csv file
        i = column                                             # Defines column numbers
        if i < 2:
            print("Incorrect Column")
        column = data[1:,i]                                    # Seperates column names and values
        converted_column =[float(i) for i in column]           # Converts values from stirngs to float
        max_value = max(converted_column)                      # Generates max value in column
        min_value = min(converted_column)                      # Generates min value in column
        avg_value = statistics.mean(converted_column)          # Generates average in column
        
        stats_table = PrettyTable()                            # PrettyTable setup for max, min and averages
        
        stats_table.field_names = ["  ","Max","Min","Avg"]
        stats_table.add_rows(
            [
                [ data[0:,i][0], max_value, min_value, avg_value]
            ]
        )
        
        print(stats_table)
        
get_column_stats(data,3)
```
